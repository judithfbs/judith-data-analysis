<h3 style="color: #120907">¿Cuál marca de carros tenemos más en inventario?</h3>

```python
dfAutos.head(3)
```

```python
dfGraf = dfAutos.groupby(['make']).count()
dfGraf = dfGraf.iloc[:,:1]
dfGraf = dfGraf.sort_values('symboling', ascending = False)
dfGraf
```

```python
fig, ax = plt.subplots(1, 1, figsize = (11,6.9), dpi = 100) #dpi, que aproveche los pixeles

eje_x = dfGraf.index.tolist()
eje_y = dfGraf['symboling'].tolist()
y_pos = np.arange(len(eje_x))

#Generación de colores para el gráfico
bar_colors = ['#000000', '#1a2546', '#465898', '#bcd3ff', '#bcd3ff',
            '#bcd3ff', '#bcd3ff', '#bcd3ff', '#bcd3ff', '#bcd3ff',
             '#bcd3ff', '#bcd3ff', '#bcd3ff', '#bcd3ff', '#bcd3ff',
             '#bcd3ff', '#bcd3ff', '#bcd3ff', '#bcd3ff', '#bcd3ff',
             '#bcd3ff', '#bcd3ff']

#---Título y subtítulo
title = "Figura 1: Cantidad de Vehículos"
ax.set_title(title,
            loc = 'left',
            x = 0.001,
            y = 0.95,
            transform = fig.transFigure,
            fontsize = 12, 
            fontweight = "bold")
            #fontfamily = xxx
    
subtitle = 'Total de vehículos agrupados según marca o fabricante'
plt.suptitle(subtitle, 
            horizontalalignment = 'left',
            x = 0.001,
            y = 0.93,
            transform = fig.transFigure,
            fontsize = 12, 
            fontweight = "regular")

#----Etiquetas en Eje X y Y
label = "Cantidad"
ax.set_xlabel(label, 
            ha = "left", 
            fontweight = "regular")

ax.xaxis.set_label_coords(x = 0,
                         y = -0.07,
                         transform = ax.transAxes)

ax.tick_params(axis = "both",
              labelsize = 12)

ax.barh(y_pos, eje_y, align = 'center', color = bar_colors)
ax.set_yticks(y_pos, labels = eje_x)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Cantidad')


plt.show()
```

<h3 style="color: #120907">¿Cuál es la distribución de los autos según estilo?</h3>
